```python
#3
swiss = data('swiss')
swiss.describe()
```

```python
#3 - 1
# I first created a function that would return any province that has a greater than 50 percent of being catholic
# This would mean majority of people would be catholic
# Then I wanted the code to return that if a country were to return True of being catholic and False if less than 50 percent
# Lastly once they return true, i wanted the boolean to run through the swiss data and return True/False for all provinces
catholic_province = swiss.Catholic >= 50
is_catholic = catholic_province == True
swiss['is_catholic'] = is_catholic
```

```python
#3 - 2
# Once I had this list I used it as a hue for my seaborn and compared catholic to fertility to do an analysis for any correlation
sns.relplot(x = 'Catholic', y = 'Fertility', data = swiss, hue = 'is_catholic')
```

```python
#3 - 3
# Education and Fertility is the strongest correlation
# The data shows the lower the education, the higher the fertility
# This more than likely correlates to the notion that lower education, typically correlates with working class
# This is also illustrated in the agriculture vs fertility data as well
# The working class has a higher fertility rate, which might be due to cultural standards that working class have more
# children to assist with agricultural work.
sns.pairplot(swiss, vars = ['Fertility', 'Agriculture', 'Examination', 'Education'])
```

```python
#4
def database(un, hst, pw, db):
    url = (f'mysql+pymysql://{user}:{password}@{host}/{db}')
    return (url)

chipotle = pd.read_sql('select * from orders', database(user, host, password, 'chipotle'))

# First thing I had to do was to remove the dollar sign from the item_price again to do the calculations in a float instead of string
# Then I created a new function called revenue that would add up all the total revenues for each menu item
# I grouped it by menu item name so that it would give me all the revenues of all the items
# Once I had them all, I used the nlargest function to only return the top 4
# I looked at the data and it showed the top 4 items with their amounts
# Lastly I used the matplotlib functions to create a bar chart out of the data I already created a table for under the top 4 function I created
# With this data, we can see that chicken bowls and burritos are going to be the 2 highest revenue products sold by chipotle
updated_price = chipotle.assign(price = chipotle['item_price'].str.replace('$','').astype(float))
revenue = updated_price.groupby('item_name').price.agg('sum')
top_4 = revenue.nlargest(4, keep = 'first')
plt.figure(figsize=(8, 6))
top_4.plot.bar(color=['red', 'royalblue', 'green', 'orange'])
plt.xlabel('Menu Item')
plt.ylabel('Revenue Generated')
plt.title('Top 4 Menu Items')
```
