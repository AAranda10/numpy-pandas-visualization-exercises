```python
import matplotlib.pyplot as plt
```

```python
##Trying out numpy
import numpy as np

x = np.linspace(0, 10, 100)
y = x**2 - x + 2  

fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
x = range(1, 100)
y = [pow(n, 2) - n + 1 for n in x]
plt.plot(x, y)
plt.title("$y = x^2 - x + 2$")
plt.annotate('Start Here', xy = (0, 0), xytext = (10, 2000), arrowprops={'facecolor': 'red'})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
import math
x = range(0, 20)
y1 = [math.sqrt(n) for n in x]
y2 = [pow(n, 3) for n in x]
y3 = [2 ** n for n in x]
y4 = [pow(n, 2) - n + 1 for n in x]

plt.plot(x, y1, c= 'red')
plt.title('Square Root 2 Function')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
plt.plot(x, y2)
plt.title('$X^3 Function$')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
plt.plot(x, y3, c= 'green')
plt.title('$(2^x) Function$')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
plt.plot(x, y4, c= 'orange')
plt.title('$(Y = X^2 - X + 2) Function$')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y1, c= 'red')

plt.subplot(2, 2, 2)
plt.plot(x, y2, c = 'royalblue')

plt.subplot(2, 2, 3)
plt.plot(x, y3, c = 'orange')

plt.subplot(2, 2, 4)
plt.plot(x, y4, c = 'purple')
```

```python
plt.plot(x, y1, c= 'red', label= 'Sqrt 2')
plt.plot(x, y2, c= 'royalblue', label= '$X^3$')
plt.plot(x, y3, c= 'orange', label= '$2^x$')
plt.plot(x, y4, c= 'green', label= '$Y = X^2 - X + 2$')

plt.xlim(0, 15)
plt.ylim(0, 30)
plt.legend(loc = 'upper right')
```
