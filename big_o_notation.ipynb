```python
import math
import matplotlib.pyplot as plt

x1 = range(1, 20)
x2 = range(0, 20)

y1 = [0*n + 1 for n in x2]
y2 = [math.log(n) for n in x1]
y3 = [n for n in x2]
y4 = [n * math.log(n) for n in x1]
y5 = [pow(n, 2) for n in x2]
y6 = [pow(2, n) for n in x2]
y7 = [math.factorial(n) for n in x1]
y8 = [pow(n, n) for n in x2]

plt.plot(y1, label= '$O(1)$')
plt.plot(y2, label= '$O(log n)$')
plt.plot(y3, label= '$O(n)$')
plt.plot(y4, label= '$O(n log n)')
plt.plot(y5, label= '$O(n^2)$')
plt.plot(y6, label= '$O(2^n)$')
plt.plot(y7, label= '$O(n!)$')
plt.plot(y8, label= '$O(n^n)$')

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.ylim(0, 20)
plt.xlim(0, 20)
plt.legend(loc = 'lower right')
```
